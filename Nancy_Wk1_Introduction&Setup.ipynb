2. venv
We will use a python virtual environment to make a self contained dev environment that has everything we need and is isolated from the global python environment.

Open the terminal in VS Code (Ctrl+`).


Create a virtual environment python -m venv env_name

Activate the virtual env env_name\Scripts\activate
(For Windows) If your folder doesn't have the required permissions for this step, you might run into SecurityError. To fix that:
   a. run cmd as administrator
   b. navigate to your folder (example cd '.\OneDrive\Desktop\Build Fellowship\')
   c. run this command Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope LocalMachine


Install required dependencies pip install -r requirements.txt

10. **Compare and contrast Git merging and rebasing. What are the advantages and potential pitfalls of using `git rebase` in a collaborative workflow?**

In Git merging and rebasing basically do the same work i.e. intergrating changes/ commits from other branches to another branch. They all work by switching into the branch in which the changes are to be directed and making it the current head branch. The difference comes in on the output of the two processes.
1. For mmerge there is a merge commit that is created which clearly shows all the hstory and gives clear picture of the two brnaches beingmerged 
2. For the rebase there is no branching out, no new commits are created, no history to be shown and it looks like the development has been done in a straight 
line.

Advantages of rebase
a) There are less commits created since history is in a straight line(linear) in the process of development hence its easier to read through the history.

Disadvantages
a) It is difficult to trace how the development process occured and at what points and from where the changes were made in the code leading to loss of previous work.
b) Can be a cause of confusion in a case where multiple developers are working on a certain project.
